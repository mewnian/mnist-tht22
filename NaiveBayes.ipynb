# Naive Bayes

This notebook uses Multinominal Naive Bayes to train and predict the mnist data

## Training

#### Importing Libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
p = np.zeros(10) 
fz = np.zeros(shape=[10,2,1600])

for i in range(60000): 
    name = str(i) 
    while len(name) < 5: 
        name = '0' + name 
    name = "digit-train\\train\\" + name
    fi = open(name + ".inp", "r") 
    fo = open(name + ".ans", "r") 
    y = int(fo.read() )
    p[y] = p[y] + 1

    tmp = fi.read()
    tmp = tmp.replace('\n', '')

    for j in range(40): 
        for k in range(40): 
            id = 40 * j + k 
            t = tmp[id]
            if t == '0': 
                fz[y][0][id] += 1
            else: 
                fz[y][1][id] += 1
```

Process P(i) - The probability of results i in dataset

```python
P = np.log(p / np.sum(p))
print(P)
```

Process fz (frequency) into lambda

```python
theta = np.zeros(shape=[10,2,1600])

for i in range(10): 
    d = 3200 
    Nn = np.sum(fz[i]) 
    theta[i] = (fz[i] + np.ones(fz[i].shape)) / (Nn + d)

theta = np.log(theta)
```

## Testing

#### Prediction Function

```python
def predict(X): 
    prob = P
    for i in range(1600):
        x = int(X[i]) 
        for j in range(10): 
            prob[j] += theta[j][x][i]
    print(prob)
    return np.argmax(prob)
```

#### Manual Testing

```python
def Testing(id): 
    name = str(id) 
    while len(name) < 5: 
        name = '0' + name 
    name = "test\\" + name
    fi = open(name + ".inp", "r") 
    fo = open(name + ".ans", "r") 
    y = int(fo.read() )
    tmp = fi.read()
    tmp = tmp.replace('\n', '')
    print("Testing Set: ", id)
    print("Predicted Value: ", predict(tmp)) 
    print("Real Value : ", y)
    print("==========================================")
```

#### Testing - Accuracy Metric - Full

```python
accuracy = 0
total = 0
for i in range(10000): 
    Testing(i)

    if y == predict(tmp): 
        accuracy += 1
    total += 1

print(accuracy / total * 100)
```
